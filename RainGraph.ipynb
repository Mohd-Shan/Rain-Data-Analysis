```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("Rainy_Data.csv")

# 1. Basic Info & Summary
print("Shape of data:", df.shape)
print("\nData Summary:")
print(df.describe())

# 2. Annual Rainfall Trend
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="YEAR", y="ANNUAL", marker='x', color='teal')
plt.title("Annual Rainfall in India (2004-2024)")
plt.ylabel("Rainfall (mm)")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Monthly Rainfall Heatmap
monthly_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUNE', 'JULY', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
monthly_matrix = df[monthly_cols].to_numpy()

plt.figure(figsize=(12, 6))
sns.heatmap(monthly_matrix.T, cmap='coolwarm', annot=True, fmt=".1f", 
            yticklabels=monthly_cols, xticklabels=df["YEAR"])
plt.title("Monthly Rainfall Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.tight_layout()
plt.show()

# 4. Seasonal Rainfall Comparison
seasonal_cols = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']
df_season = df[['YEAR'] + seasonal_cols]

plt.figure(figsize=(12, 6))
df_season.set_index('YEAR')[seasonal_cols].plot(kind='bar', stacked=True, colormap='Accent', figsize=(12, 6))
plt.title("Seasonal Rainfall in India")
plt.ylabel("Rainfall (mm)")
plt.xlabel("Year")
plt.legend(title="Season")
plt.tight_layout()
plt.grid(axis='y')
plt.show()

# 5. NumPy Stats
annual_array = df["ANNUAL"].to_numpy()
print("\nNumPy Stats:")
print("Mean Annual Rainfall:", np.mean(annual_array))
print("Standard Deviation:", np.std(annual_array))
print("Maximum:", np.max(annual_array))
print("Minimum:", np.min(annual_array))
```

```python
import matplotlib.pyplot as plt

# Scatter plot: Year vs Annual Rainfall
plt.figure(figsize=(10, 6))
plt.scatter(df['YEAR'], df['ANNUAL'], color='green', marker='x')
plt.title('Annual Rainfall in India (20042008)')
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.grid(True)
plt.show()
```
